```python
import os
os.makedirs("../data/text_files", exist_ok=True)
```

```python
sample_text = {"../data/text_files/sample1.txt":"""Language models (LMs) have recently been augmented with retrieval mechanisms to enhance their performance on knowledge-intensive tasks. These retrieval-augmented language models (RALMs) leverage external knowledge sources, such as large text corpora or databases, to provide relevant information during the generation process. By incorporating retrieval capabilities, RALMs can access up-to-date and domain-specific information, leading to improved accuracy and relevance in their responses. This approach has shown promise in various applications, including question answering, dialogue systems, and content generation, where access to external knowledge is crucial for generating informative and contextually appropriate outputs."""}
for file_path, text in sample_text.items():
    with open(file_path, "w", encoding="utf-8") as f:
        f.write(text)

print("text file created")

```
